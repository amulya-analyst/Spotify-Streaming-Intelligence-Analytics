```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between Spotify Features")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="popularity", y="streams", hue="genre")
plt.title("Popularity vs Streams by Genre")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['genre'].value_counts().plot(kind='bar')
plt.title("Number of Songs by Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="genre", y="streams")
plt.xticks(rotation=45)
plt.title("Average Streams by Genre")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["popularity"], bins=20, kde=True)
plt.title("Popularity Distribution")
plt.show()
```

```python
df["duration_min"] = df["duration_ms"] / 60000
plt.figure(figsize=(8,5))
sns.histplot(df["duration_min"], bins=20)
plt.title("Song Duration Distribution (Minutes)")
plt.show()
```

```python
sns.barplot(data=df, x="subscription_plan", y="revenue_usd")
plt.title("Revenue by Subscription Plan")
plt.show()
```

```python
country_streams = df.groupby("country")["streams"].sum().sort_values(ascending=False)

country_streams.plot(kind="bar")
plt.title("Total Streams by Country")
plt.show()
```

```python
features = ['danceability','energy','loudness','tempo','valence','popularity']

sns.pairplot(df[features], corner=True)
plt.show()
```

## Key Insights
1. The dataset contains no missing values, indicating clean and complete data.

2. Energy and loudness show a strong positive correlation, suggesting louder songs tend to have higher energy.

3. Popularity scores do not show a strong correlation with streaming counts, indicating that streams may depend on other factors.

4. Genre analysis shows variation in average streaming performance across music categories.

5. Most songs fall within the mid-range popularity scores, suggesting that extremely popular songs are relatively rare.
